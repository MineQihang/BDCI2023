```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results_all = pd.read_csv('./submit/csv/test_results_rough_lgb.csv')
results_partial_30 = pd.read_csv('./submit/csv/test_results_11_12_partial_30.csv')
results_partial_90 = pd.read_csv('./submit/csv/test_results_11_12_partial_90.csv')
results_partial_365 = pd.read_csv('./submit/csv/test_results_11_12_partial_365.csv')
```

```python

plt.plot(results_partial_30['x_k'], label='30')
plt.plot(results_partial_90['x_k'], label='90')
plt.plot(results_partial_365['x_k'], label='365')
plt.plot(results_all['x_k'], label='all')
plt.legend()
plt.show()
```

```python
# 将所有的结果合并
results_all['x_k'] = np.max([results_all['x_k'], results_partial_30['x_k'], results_partial_90['x_k'], results_partial_365['x_k']], axis=0)
results_all.to_csv('./submit/csv/test_results_11_12_gather.csv', index=False)
```
